# Задание 1

```python
data = pd.read_csv('ratings.csv')[['userId', 'timestamp']]
```

```python
a = data.userId.value_counts()
```

```python
data_userId_count_more_100 = a[a > 100]
```

```python
data_userId_count_more_100 = pd.DataFrame(data_userId_count_more_100).reset_index()
```

```python
data_userId_count_more_100 = data_userId_count_more_100[['index']]
```

```python
data_userId_count_more_100.columns = ['userId']
```

```python
data_userId_count_more_100.sort_values(by = 'userId', inplace = True)
```

```python
data_userId_count_more_100.index = [i for i in range(len(data_userId_count_more_100))]
```

```python
time_life = data.groupby('userId').max() - data.groupby('userId').min()
```

```python
time_life = time_life[['userId', 'timestamp']]
time_life
```

```python
data_userId_count_more_100.merge(time_life, how = 'inner')
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
base1 = client_base[['client_id']].merge(rzd, how = 'left').merge(auto, how = 'left').merge(air, how = 'left').fillna(0)
base1
```

```python
base2 = client_base.merge(base1, how = 'outer')
base2
```
